Girilen sayının çift mi tek mi olduğunu döndüren fonksiyon yazınız.

````python
def tek_cift(sayi):
  if sayi % 2 == 0:
    return "Çift"
  else:
    return "Tek"
````

# Soru 3:  Alışveriş İndirimi

Alışveriş tutarına göre indirim uygulayan fonksiyon yazınız.

* 1000 tl üzerinde %20

* 500-1000 arasında %10

* 500 tl altı indirim yok.



```
Not: Fonksiyonun girdisi olmasın, fonksiyon içerisinde kullanıcıdan tutar girmesi istensin.
```

```
Not 2: Müşterinin ödeyeceği tutar çıktı olarak yazdırılsın.
```





````python
def hesapla():
  tutar = float(input("Alışveriş tutarını giriniz: "))
  if tutar > 1000:
    indirim = tutar * 0.2
    odenecek = tutar - indirim
    print("indirim uygulanan tutar: ", indirim)
  elif tutar >= 500:
    indirim = tutar * 0.1
    odenecek = tutar - indirim
    print("indirim uygulanan tutar: ", indirim)
  else:
    odenecek = tutar
    print("Malesef indirim kazanamadınız.")

  print("Ödemeniz gereken tutar: ", odenecek)

  return odenecek

````

````python
hesapla()
````

# Soru 4:  Şifre Güvenlik Kontrolü

Şifrenin güvenli olup olmadığını kontrol eden fonksiyon yazınız.

Kurallar:

* en az 6 karakter içermelidir.

* en az 1 rakam içermelidir.





```
Not: Şifre 6 haneden küçükse "Uyarı Mesajı" versin.
```

```
Not 2: Şifre 6 haneden büyükse ve rakam içeriyorsa "Güçlü Şifre" rakam içermiyorsa "Zayıf Şifre" çıktılarını döndürsün.
```





````python
def sifre_kontrol(sifre):
  if len(sifre) < 6:
    print("Şifreniz en az 6 karakter olmalıdır.")

  rakam_var = False
  for i in sifre:
    if i.isdigit():
      rakam_var = True
      break

  if rakam_var:
    print("Güçlü Şifre")
  else:
    print("Zayıf Şifre")
````

````python
sifre_kontrol("nggjkgd")
````
