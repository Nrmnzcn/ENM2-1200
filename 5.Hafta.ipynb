## Fonksiyon Yapısı





Python'da fonksiyon tanımlamak, belirli bir görevi yerine getiren kod bloklarını organize etmenin ve yeniden kullanmanın temel yoludur. Bir fonksiyonun temel yapısı şu bileşenlerden oluşur:

-  **`def` anahtar kelimesi:** Fonksiyon tanımının başladığını belirtir. Her fonksiyon tanımı `def` ile başlamalıdır.
- **Fonksiyon adı:** Fonksiyonu çağırmak için kullanılan benzersiz bir isimdir. Fonksiyon adları genellikle küçük harflerle başlar ve kelimeler arasında alt çizgi (`_`) kullanılır (örneğin, `fonksiyon_adi`).
-  **Parametreler:** Fonksiyonun çalışması için gereken girdilerdir. Fonksiyon adından sonra parantez `()` içinde belirtilir. Birden fazla parametre virgülle ayrılır. Parametreler isteğe bağlıdır; yani bir fonksiyon hiç parametre almayabilir (`def fonksiyon_adi():`).
-  **İki nokta üst üste (`:`):** Fonksiyon başlığının sonunu ve fonksiyon gövdesinin başlangıcını belirtir.
-  **Girintileme (Indentation):** Fonksiyon gövdesindeki tüm kod satırları aynı miktarda girintili (genellikle 4 boşluk) olmalıdır. Python'da girintileme, kod bloklarının kapsamını (hangi kodun fonksiyona ait olduğunu) belirler ve sözdiziminin önemli bir parçasıdır.
- **`return` ifadesi:** Fonksiyonun bir değer üretip bu değeri çağıran yere geri göndermesini sağlar. `return` ifadesi isteğe bağlıdır; bir fonksiyon açıkça bir değer döndürmezse, varsayılan olarak `None` döndürür.

## Örnek 2: Biyomedikal Alan

* Problem:
Hastanın Vücut Kitle İndeksi (BMI) hesaplama

Formül:
`BMI = kilo / (boy^2)`

````python
def bmi_hesapla(kilo, boy):
    return kilo / (boy ** 2)

def bmi_yorumla(bmi):
    if bmi < 18.5:
        return "Zayıf"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Fazla Kilolu"
    else:
        return "Obez"
````

````python
bmi_nermin = bmi_hesapla(70, 1.68)
bmi_yorumla(bmi_nermin)
````

````python
bmi = bmi_hesapla(89, 1.62)
bmi_yorumla(bmi)
````

## Örnek 3: İçerisinde Döngü bulunduran Fonksiyonlar

İlk **N** sayının toplamı hesaplayabileceğimiz bir fonksiyon yazınız.

````python
def ilk_n_toplam(n):
    toplam = 0
    for i in range(1, n+1):
        toplam += i
    return toplam
````

````python
print(ilk_n_toplam(9))
````

````python
print(ilk_n_toplam(156))
````

## Örnek 4: Endüstri Mühendisliği Örneği

Kalite Kontrol Uyarısı veren bir fonksiyon yazalım.

````python
def uretim_verimliligi_hesapla(uretilen_urun_adedi: int, hedef_urun_adedi: int, ariza_adedi: int):
    """
    Endüstriyel üretim verimliliğini hesaplar.

    Parametreler:
    uretilen_urun_adedi (int): Gerçekte üretilen toplam ürün adedi.
    hedef_urun_adedi (int): Hedeflenen üretim adedi.
    ariza_adedi (int): Üretilen ürünler arasındaki arızalı ürün adedi.

    Çıktı: Verimlilik oranı ve bir durum mesajı.
    """

    # Sıfıra bölme hatasını engellemek için koşul ifadesi
    if hedef_urun_adedi <= 0:
        return 0.0, "Hata: Hedeflenen ürün adedi sıfır veya negatif olamaz."

    # Üretilen ürün adedinin hedefin altında veya arızalı ürün adedinden az olmaması kontrolü
    if uretilen_urun_adedi < 0 or ariza_adedi < 0:
        return 0.0, "Hata: Ürün veya arıza adedi negatif olamaz."

    if ariza_adedi > uretilen_urun_adedi:
        return 0.0, "Hata: Arızalı ürün adedi, üretilen ürün adedinden fazla olamaz."

    # Net üretim adedi
    net_uretilen_urun = uretilen_urun_adedi - ariza_adedi

    # Verimlilik oranı hesaplaması
    verimlilik_orani = (net_uretilen_urun / hedef_urun_adedi) * 100

    # Durum mesajı için koşul ifadeleri
    durum_mesaji = ""
    if verimlilik_orani >= 90:
        durum_mesaji = "Yüksek verimli: Üretim hedeflerinin üzerinde veya yakınında."
    elif 70 <= verimlilik_orani < 90:
        durum_mesaji = "Ortalama verimli: Üretim hedeflerine ulaşmak için iyileştirme gerekebilir."
    else:
        durum_mesaji = "Düşük verimli: Ciddi iyileştirme gerekmekte."

    return verimlilik_orani, durum_mesaji
````

````python
verimlilik1, durum_mesaji1 = uretim_verimliligi_hesapla(uretilen_urun_adedi=1000, hedef_urun_adedi=1000, ariza_adedi=20)
print(f"Verimlilik Oranı: {verimlilik1:.2f}%\nDurum Mesajı: {durum_mesaji1}")
````

````python
verimlilik2, durum_mesaji2 = uretim_verimliligi_hesapla(uretilen_urun_adedi=850, hedef_urun_adedi=1000, ariza_adedi=50)
print(f"Verimlilik Oranı: {verimlilik2:.2f}%\nDurum Mesajı: {durum_mesaji2}")
````

````python
verimlilik3, durum_mesaji3 = uretim_verimliligi_hesapla(uretilen_urun_adedi=100, hedef_urun_adedi=0, ariza_adedi=5)
print(f"Verimlilik Oranı: {verimlilik3:.2f}%\nDurum Mesajı: {durum_mesaji3}")
````

````python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
````
