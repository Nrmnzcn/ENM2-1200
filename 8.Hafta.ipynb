## Liste Dilimleme (Slicing)





Listeden bir aralığı seçmek, kesmek anlamına gelir.

> [index] -> o sıradaki elemanı

> [baslangic : bitis] -> belirtilen indesler arasındaki elemanları

> [baslangic : bitis] -> belirtilen indesler arasındaki elemanları

> [baslangic : bitis : adım] -> belirtilen indesler arasındaki elemanları belirtilen adım sayısında atlayarak

seçer


````python
rakamlar = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]

print(rakamlar[5])
print(rakamlar[1:3])
````

### **Not:** Bitiş indeksine kadar dilimlenir, son eleman liseteye dahil edilmez.

### **Not:**

> [baslangic : bitis : adım] -> Bazen başlangiç ve bitiş indexleri verilmez, bu listenin başından başla veya listenin sonuna kadar devam et anlamı taşır.

````python
rakamlar = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]

print(rakamlar[:5])
````

````python
print(rakamlar[2:])
````

````python
print(rakamlar[::2])
````

## Liste Elemanlarını Değiştirme

Verilen indexe yeni bir değer atanabilir. Eski değerin yerini yeni atanan değer alır.

````python
notlar = [70, 85, 90, 60]

notlar[0] = 99

print(notlar)
````

## Liste Metotları

(Listeler ile birlikte kullanılabilen fonksiyonlar)

### append()

Listenin sonuna verilen veriyi ekler.


````python
notlar.append(72)
print(notlar)
````

````python
notlar.append([45, 56])
print(notlar)
````

````python
len(notlar)
````

### insert(index, yeni değer)

Verilen indexe yeni değer ekler. O indexteki değer kaybolmaz, araya ekleme yapar.

````python
notlar.insert(1, 80)
print(notlar)
````

### remove()

Listedeki belirtilen elemanı listeden kaldırır.


````python
notlar.remove(85)
print(notlar)
````

````python
notlar.remove(0)
print(notlar)
````

### pop()

Listeden belirtilen indexteki elemanı kaldırır. Default değeri -1'dir yani index verilmezse son elemanı kaldırır.



````python
notlar.pop(1)
print(notlar)
````

````python
notlar.pop()
print(notlar)
````

### sort()

Listedeki elemanları sıralar.

````python
sayilar = [5, 2, 9, 1]
sayilar.sort()
print(sayilar)
````

````python
sehirler = ["Hatay", "Kahraman Maraş", "Adana"]
sehirler.sort()
print(sehirler)
````

````python
karma_liste = [10, "Python", 3.14, True, ["Nermin", 59]]
karma_liste.sort()
print(karma_liste)
````

### reverse()

Listeyi sondan başa sıralar, tersine çevirir.

````python
rakamlar = [1, 2, 3, 4, 5, 6, 7, 8, 9]

rakamlar.reverse()
print(rakamlar)
````

## Örnek Not Hesabı

````python
notlar = [70, 85, 90, 60]

toplam = 0

for n in notlar:
    toplam += n

ortalama = toplam / len(notlar)

print("Ortalama:", ortalama)
````

## Listedeki En Büyük Sayıyı Bulma

````python
sayilar = [3, 10, -196, 7, 25, 1, 98, -77]

max_sayi = sayilar[0]

for s in sayilar:
    if s > max_sayi:
        max_sayi = s

print("En büyük sayı:", max_sayi)
````

## Listedeki Belirli Değerleri Sayma

````python
sayilar = [1, 2, 3, 2, 2, 4, 8]

sayac = 0

for s in sayilar:
    if s == 2:
        sayac += 1

print("2 sayısı:", sayac, "kez var")
````

#### Fonksiyon olarak

````python
def saydir(liste, deger):

  sayac = 0

  for s in liste:
    if s == deger:
      sayac += 1

  print(f"{deger} sayısı:", sayac, "kez var")

  return
````

````python
sayilar = [1, 2, 3, 2, 2, 4, 8]
saydir(sayilar, 2)
````

````python
saydir(sayilar, 9)
````

### count() fonksiyonu

````python
sayilar.count(9)
````
