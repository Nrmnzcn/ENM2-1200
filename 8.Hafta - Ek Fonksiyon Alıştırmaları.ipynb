### Soru 1:


> Ahmet bugün 20 yaşındadır. 6 yıl önce Ahmet’in yaşı, kardeşi Mehmet’in yaşının 2 katıydı. Mehmet bugün kaç yaşındadır?





Bu problemi çözen bir fonksiyon yazınız

📌 Fonksiyon:

* Ahmet’in bugünkü yaşını girdi olarak alsın,
* Mehmet’in yaşını **döndürsün**.

````python
def mehmet_yas_hesapla(ahmet_yas):

  a_6_once = ahmet_yas - 6
  m_6_once = a_6_once / 2
  m_bugun = m_6_once + 6

  return m_bugun
````

````python
mehmet_yas_hesapla(20)
````

### Soru 2:


> Bir baba ile oğlunun yaşları toplamı 50’dir. Baba, oğlunun yaşının 4 katıdır. Baba ve oğlu kaç yaşındadır?





Bu problemi çözen bir fonksiyon yazınız

📌 Fonksiyon:

* Toplam yaşı girdi olarak alsın.
* Baba ve oğul yaşlarını **fonksiyon içerisinde yazdırsın**.

````python
def hesapla(toplam_yas):

  cocuk_yası = toplam_yas / 5
  baba_yası = cocuk_yası * 4
  print(f"Baba {int(baba_yası)} yaşındadır.")
  print(f"Oğul {int(cocuk_yası)} yaşındadır.")

  return
````

````python
hesapla(50)
````

### Soru 3:

> Bir kargo firması gönderim ücretini hem ağırlık (kg) hem de hacim (desi) bilgisine göre hesaplamaktadır. Kargo firmaları genellikle hangisi daha büyükse (ağırlık veya hacim) ona göre ücret belirler.



```
Ücret Tablosu:
0 – 5 arası → 60 TL
5 – 10 arası → 100 TL
10 üzeri → 150 TL
```

> Ek Ücret: Eğer gönderi şehir dışına gidiyorsa → +30 TL eklenir.


Bu problemi çözen bir fonksiyon yazınız

📌 Fonksiyon:

* agirlik, hacim, sehir_disi bilgilerini girdi olarak alsın,
* toplam kargo ücretini çıktı olarak hesaplasın.

````python
def kargo_hesaplama(agirlik, hacim, sehir_disi="Evet"):

  faturalandirma = max(agirlik, hacim)

  if faturalandirma <=5:
    odeme = 60
  elif faturalandirma <=10:
    odeme = 100
  else:
    odeme = 150

  if sehir_disi == "Evet":
    odeme += 30
  else:
    odeme += 0

  return odeme
````

````python
kargo_hesaplama(2, 7, sehir_disi="Evet")
````

````python
kargo_hesaplama(12, 4, sehir_disi="Hayır")
````

### Soru 4:

> Bir müşterinin alışveriş sepetindeki ürün fiyatları aşağıda bir liste olarak verilmektedir.

```
fiyatlar = [120, 80, 300, 50, 200]
```

> Kullandığı e-ticaret sisteminin 100 TL üzerindeki ürünlere %15 indirim uyguladığı bilinmektedir.

Bu problem için bir fonksiyon yazınız

📌 Fonksiyon:

* Aldığı ürünlerin listesi girdi olarak alsın,
* Toplam ödenecek miktarı çıktı olarak hesaplasın.

````python
def indirimli_ödeme(fiyatlar):

  indirimli_liste=[]

  for i in fiyatlar:
    if i > 100:
      indirim = i * 0.15
      indirimli = i - indirim
      indirimli_liste.append(indirimli)
    else:
      indirimli_liste.append(i)

  print(indirimli_liste)
  toplam = sum(indirimli_liste)
  return toplam
````

````python
fiyatlar = [120, 80, 300, 50, 200]
indirimli_ödeme(fiyatlar)
````
