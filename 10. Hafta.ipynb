## Yeni Sütun Üretme

````python
df["family_size"] = df.SibSp + df.Parch +1
df.sample(7)
````

````python
df["child"]=(df.Age<18)
df
````

````python
df["fare_group"]= df.Fare.apply(lambda x:"High" if x>50 else "Low")
df.sample(6)
````
