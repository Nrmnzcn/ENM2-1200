# Veri Yapıları II – Sözlükler (Dictionaries)

Bu derste öğrenciler:

- Sözlük (dictionary) veri yapısını öğrenir,
- Anahtar-değer (key-value) mantığını kavrar,
- Gerçek hayat problemlerinde sözlük kullanımını uygular,
- Veri erişimi, güncelleme ve döngüler ile kullanımını öğrenir.

**Sözlük (Dictionary) Nedir?**

> Sözlükler, anahtar (key) – değer (value) mantığıyla çalışan veri yapılarıdır.

📌 Değişken atama → x=5

📌 Listeler → [0,1,2] → index

📌 Tüpler → (0,1,2) → index

📌 Sözlüklerde ise anahtar vardır → "isim", "yas"

## Sözlük Oluşturma

Boş sözlük:

```
bos = {}
```




````python
ogrenci = {
    "isim": "Ali",
    "yas": 20,
    "bolum": "Bilgisayar Mühendisliği"
}
````

````python
print(ogrenci["isim"])
print(ogrenci["yas"])
````

## keys() and values()

````python
print(ogrenci.keys())
````

````python
print(ogrenci.values())
````

## Değer Güncelleme

````python
ogrenci["yas"] = 26
print(ogrenci)
````

## Yeni Eleman Ekleme

````python
ogrenci["sehir"] = "Ankara"
print(ogrenci)
````

## Eleman Silme

````python
del ogrenci["sehir"]
````

````python
# veya
ogrenci.pop("yas")
````

````python
print(ogrenci)
````

## Döngü kullanımı

````python
kisi = {
    "ad": "Ayşe",
    "yas": 25,
    "sehir": "İstanbul"
}
````

````python
# Sadece anahtarlar
for key in kisi:
    print(key)
````

````python
# Anahtar + değer

for key, value in kisi.items():
    print(key, ":", value)
````

## Sözlük Uzunluğu

````python
print(len(kisi))
````

## İç İçe Sözlükler (Nested Dictionary)

````python
ogrenciler = {
    "ogrenci1": {
        "isim": "Ali",
        "yas": 20
    },
    "ogrenci2": {
        "isim": "Ayşe",
        "yas": 22
    }
}
````

````python
len(ogrenciler)
````

## Liste + Sözlük Birlikte Kullanım

````python
sinif = [
    {"isim": "Ali", "not": 80},
    {"isim": "Ayşe", "not": 90},
    {"isim": "Mehmet", "not": 70}
]

for ogrenci in sinif:
    print(ogrenci["isim"], ":", ogrenci["not"])
````

# Örnek 1: Öğrenci Bilgi Sistemi

````python
ogrenci = {}

ogrenci["isim"] = input("İsim: ")
ogrenci["yas"] = int(input("Yaş: "))
ogrenci["not"] = float(input("Not: "))

print("\n--- Öğrenci Bilgisi ---")
for key, value in ogrenci.items():
    print(key, ":", value)
````

# Örnek 2: Not Hesaplama Sistemi

````python
notlar = {
    "Ali": 85,
    "Ayşe": 92,
    "Mehmet": 70
}

toplam = 0

for not_degeri in notlar.values():
    toplam += not_degeri

ortalama = toplam / len(notlar)

print("Ortalama:", ortalama)
````

# Örnek 3: En Yüksek Notu Bulma

````python
notlar = {
    "Ali": 85,
    "Ayşe": 92,
    "Mehmet": 70
}

en_yuksek = 0
en_iyi = ""

for isim, notu in notlar.items():
    if notu > en_yuksek:
        en_yuksek = notu
        en_iyi = isim

print("En başarılı öğrenci:", en_iyi)
````
